```python
data = pd.read_csv('clean_adult.csv').drop("Unnamed: 0", axis=1)
data
```

```python
data["workclass"] = data["workclass"].map({ 
    #GOV
    "State-gov":0,
    "Federal-gov":1,
    "Local-gov":2,
    #Private or Self Employed
    "Private":3,
    "Self-emp-not-inc":4,
    "Self-emp-inc":5,
    "Without-pay":6
    })
```

```python
data["marital-status"] = data["marital-status"].map({
    # Married
    "Married-civ-spouse":0,
    "Married-AF-spouse":1,
    "Married-spouse-absent":2,
    # Once Married
    "Divorced":3,
    "Separated":4,
    "Widowed":5,
    #Never Married
    "Never-married":6})
```

```python
data["occupation"] = data["occupation"].map({
    #Manual work
    "Transport-moving":0,
    "Priv-house-serv":1,
    "Machine-op-inspct":2,
    "Handlers-cleaners":3,
    "Farming-fishing":4,
    "Craft-repair":5,
    #Desk work
    "Adm-clerical":6,
    "Exec-managerial":7,
    "Prof-specialty":8,
    "Sales":9,
    "Tech-support":10,
    #Armed Forces
    "Armed-Forces":11,
    "Protective-serv":12,
    # Other
    "Other-service":13})
```

```python
data["race"] = data["race"].map({
    "White":0,
    'Black':1,
    'Asian-Pac-Islander':2,
    'Amer-Indian-Eskimo':3,
    'Other':4})
```

```python
data["sex"] = data["sex"].map({
    "Male":0,
    "Female":1})
```

```python
data = data[data["native-country"] != 'South']
```

```python
data["native-country"] = data["native-country"].map({
    #North America
    "United-States":0,
    "Canada":1,
    'Outlying-US(Guam-USVI-etc)':2,
    "Puerto-Rico":3,
    #South/Central America 
    "Cuba":4,
    "Jamaica":5,
    "Mexico":6,
    "Honduras":7,
    "Columbia":8,
    "Ecuador":9,
    "Haiti":10,
    "Dominican-Republic":11,
    "El-Salvador":12,
    "Guatemala":13,
    "Peru":14,
    'Trinadad&Tobago':15,
    "Nicaragua":16,
    #ASIA
    "India":17,
    "Philippines":18,
    "Cambodia":19,
    "Thailand":20,
    "Laos":21,
    "Taiwan":22,
    "China":23,
    "Japan":24,
    "Vietnam":25,
    "Hong":26,
    #Europe
    "England":27,
    "Germany":28,
    "Poland":29,
    "Portugal":30,
    "France":31,
    "Italy":32,
    "Yugoslavia":33,
    "Scotland":34,
    "Greece":35,
    "Ireland":36,
    "Hungary":37,
    "Holand-Netherlands":38,
    #Middle East
    "Iran":39
    })
```
