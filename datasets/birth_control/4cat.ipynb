```python
import pandas as pd
```

```python
data = pd.read_csv("cmc.csv", names=["age","wife_ed","husb_ed","no_kids","wife_rel","wife_works",
    "husb_occupation","SOL_index","media_exp","class"])
```

```python
data
```

```python
ages = pd.cut(data["age"], [16,19,29,39,49], labels=[0,1,2,3], include_lowest=True)
data["age"] = ages
```

```python
kids = pd.cut(data["no_kids"], [0,0.5,4,10,16], labels=[0,1,2,3], include_lowest=True)
data["no_kids"] = kids
```

```python
data["husb_occupation"] = data["husb_occupation"].map(lambda x: x-1)
```

```python
for c in data.columns:
    print(c, set(data[c]))
```

```python
data.to_csv("cmc_4cat.csv", index=False, header=False)
```
