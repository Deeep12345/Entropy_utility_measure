```python
import pandas as pd
```

```python
df = pd.read_csv("framingham_heart_disease.csv")
```

```python
df.rename(columns={"male":"sex"}, inplace=True)
```

```python
df.isnull().sum()
```

```python
for c in df.columns:
    print(c, set(df[c]))
```

```python
df = df.dropna(axis=0)
df
```

```python
for c in df.columns:
    if len(set(df[c])) > 20:
        col = df[c]
        print(c)
        print(min(col), max(col))
        newc, bins = pd.cut(df[c], 20, labels=[i for i in range(20)], retbins=True)
        print(bins)
        df[c] = newc
```

```python
df
```

```python
df["education"] = df["education"].map(lambda x: int(x))
df["BPMeds"] = df["BPMeds"].map(lambda x: int(x))
for c in df.columns:
    print(c, set(df[c]))
```

```python
for c in df.columns:
    vals = set(df[c])
    print(c, vals)
    m = {v:i for i, v in enumerate(vals)}
    df[c] = df[c].map(m)
    print(set(df[c]))
```

```python
for c in df.columns:
    print(c, set(df[c]))
```

```python
df.to_csv("heart_cat.csv", index=None)
```
