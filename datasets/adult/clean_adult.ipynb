```python
import pandas as pd
```

```python
cols = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", 
        "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
       "hours-per-week", "native-country", "class"]
df = pd.read_csv("adult.csv", names=cols, index_col=False, delimiter=", ")
df
```

```python
df["class"] = df["class"].map({"<=50K": 0, ">50K": 1})
df = df.drop(["fnlwgt", "education"], axis=1)
df
```

```python
cat_attributes = df.select_dtypes(include=['object'])
print(cat_attributes.columns)
mapping = {}
for c in cat_attributes.columns:
    vals = set(df[c])
    no_vals = len(vals)
    m = {}
    for i, v in enumerate(vals):
        m[v] = i
        
    mapping[c] = m
    df[c] = df[c].map(m)
    
print(mapping)
df
```

```python
num_attributes = df.select_dtypes(include=['int'])
print(num_attributes.columns)

for c in num_attributes.columns:
    if len(set(df[c])) > 20:
        col = df[c]
        print(c)
        print(min(col), max(col))
        newc, bins = pd.cut(df[c], 20, labels=[i for i in range(20)], retbins=True)
        print(bins)
        df[c] = newc

df
        
```

```python
no_oh_cols = 0
for c in df.columns[:-1]:
    vals = set(df[c])
    no_cols = max(vals) + 1 - min(vals)
    print(c, no_cols)
    no_oh_cols += no_cols
print(no_oh_cols)
    
```

```python
df["education-num"] = df["education-num"].map(lambda x: x-1)
```
