```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data = pd.read_csv("results.csv")
data
```

```python
plt.plot(data.k, data.accuracy)
plt.title("Accuracy for changing Ks")
plt.xlabel("K")
plt.ylabel("Training accuracy")
plt.show()
```

```python
plt.plot(data.k, data.entropy)
plt.title("Entropy for changing Ks")
plt.xlabel("K")
plt.ylabel("Relative entropy of data")
plt.show()
```

```python
fig, ax = plt.subplots()

plt.plot(data.entropy, data.accuracy)
ax.set_xlim(1, 0)  # decreasing time
plt.title("Utility vs Entropy")
plt.xlabel("Entropy")
plt.ylabel("Accuracy")
plt.show()
```

```python
ax = data.plot(x="k", secondary_y=["no_classes"], figsize=(15,10))
ax.set_ylabel("Accuracy and Rel. Entropy")
ax.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
ax.right_ax.legend(loc='center left', bbox_to_anchor=(1.1, 0.4))
ax.right_ax.set_ylabel("Number of EQ classes")

plt.show()
```
