```python
import pandas as pd
```

```python
cols = ["age","workclass","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","class"]
data = pd.read_csv("../../../datasets/adult/adult_cat_mapped.csv", names=cols)
```

```python
oh = data[cols[:-1]]
oh
```

```python
oh = pd.get_dummies(oh, columns=cols[:-1])
```

```python
oh["class"] = data["class"]
```

```python
oh
```

```python
for c in cols:
    rel_cols = list(filter(lambda x: c == x[:len(c)], oh.columns))
    print(rel_cols)
    res = oh.groupby(rel_cols).size()
    print(c, res.index)
```

```python
m = {}
for c in cols:
    if c == "class":
        m[c] = c
    else:
        rel_cols = list(filter(lambda x: c == x[:len(c)], oh.columns))
        for i in range(len(rel_cols)):
            m[c+f"_{i}"] = c+str(i)
m
        
```

```python
oh = oh.rename(m, axis=1)
```

```python
oh.to_csv("original_oh.csv", index=False)
```
