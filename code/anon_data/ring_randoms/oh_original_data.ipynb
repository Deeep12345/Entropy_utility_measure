```python
cols = [f"{i}th" for i in range(20)]
cols.append("class")
data = pd.read_csv("../../../datasets/ring/ring_cat.csv", names=cols)
```

```python
for c in [f"{i}th" for i in range(20)]:
    rel_cols = list(filter(lambda x: c == x[:len(c)], oh.columns))
    print(rel_cols)
    res = oh.groupby(rel_cols).size()
    print(c, res.index)
```

```python
m = {}
for c in cols:
    if c == "class":
        m[c] = c

    for i in range(20):
        m[c+f"_{i}"] = c+str(i)
m
        
```

```python
oh = oh.rename(m, axis=1)
```

```python
oh.to_csv("original_oh.csv", index=False)
```
