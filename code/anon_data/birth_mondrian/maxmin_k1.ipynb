```python
import pandas as pd
```

```python
data = pd.read_csv("k1.csv", names=["age","wife_ed","husb_ed","no_kids","wife_rel","wife_works",
    "husb_occupation","SOL_index","media_exp","class"])
```

```python
newd = data
for col in data.columns[:-1]:
    newd[f"{col}_min"] = data[col]
    newd[f"{col}_max"] = data[col]
    newd = newd.drop(col, axis=1)
    
y = newd["class"]
newd = newd.drop("class", axis=1)
newd["class"] = y
newd
```

```python
newd.to_csv("k1_minmaxed.csv")
```
