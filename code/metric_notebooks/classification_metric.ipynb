```python
import numpy as np
import pandas as pd

import re
import time
```

```python
anon_data = pd.read_csv('anon_data/ring_mondrian/k7400_minmaxed.csv')
anon_data.rename(columns={"Unnamed: 0":"index"}, inplace=True)
anon_data.set_index("index")
```

```python
print(list(anon_data.columns[1:-1]))
grouped = anon_data.groupby(list(anon_data.columns[1:-1]))
```

```python
tot = 0

for _, classes in grouped:
    cs = classes["class"].value_counts()
    print(cs)
    max_class = cs.idxmax()
    print(f"\tThis is the max class: {max_class}")
    
    pen_eq = cs.sum() - cs[max_class]
    print(f"\tPenalty for class: {pen_eq}")
    print(f"\n\n")
    
    tot+= pen_eq
    
print("#########################################")
print(f"Total classification penalty: {tot}")
print(f"Classification Metric: {tot/len(anon_data)}")
```
