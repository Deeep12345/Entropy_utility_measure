```python
import numpy as np
import pandas as pd

import re
import time
```

```python
anon_data = pd.read_csv('anon_data/ring_mondrian/k15_minmaxed.csv')
anon_data.rename(columns={"Unnamed: 0":"index"}, inplace=True)
anon_data.set_index("index")
```

```python
cols = list(anon_data.columns[1:-1])

def distance(u,v):
    u = u[cols]
    v = v[cols]
    d = u == v
    counts = d.value_counts()
    return counts[False] if False in counts else 0
```

```python
def diam_m(data):
    data = data[cols]
    data = data.drop_duplicates()
    diam = 0
    for x in range(len(data)):
        for y in range(x+1, len(data)):
            
            d = distance(data.iloc[x], data.iloc[y])
            
            if d == len(cols):
                print(f"max distance is {d}")
                return d
            
            if d > diam:
                diam = d
                
    return diam

diam_m(anon_data)
```
