```python
import pandas as pd
import numpy as np
```

```python
births = pd.read_csv("../birth_randoms/metrics.csv").drop(["precision", "lr_acc", "auroc"], axis=1)
rings = pd.read_csv("../ring_randoms/metrics.csv").drop(["precision", "lr_acc", "auroc"], axis=1)
adults = pd.read_csv("../adult_randoms/metrics.csv").drop(["precision", "lr_acc", "auroc"], axis=1)
hearts = pd.read_csv("../heart_randoms/metrics.csv").drop(["precision", "lr_acc", "auroc"], axis=1)
births
```

```python
births_accs = pd.read_csv("../birth_randoms/accuracy_results.csv")
rings_accs = pd.read_csv("../ring_randoms/accuracy_results.csv")
adults_accs = pd.read_csv("../adult_randoms/accuracy_results.csv")
hearts_accs = pd.read_csv("../heart_randoms/accuracy_results.csv")
births_accs
```

```python
from random import randint

def gencoordinates(n, size):
    pairs = set()
    while len(pairs) < size:
        x, y = randint(1,n),randint(1,n)
        if (x,y) not in pairs and (y,x) not in pairs and x != y:
            pairs.add((x,y))
    return pairs
            
#gencoordinates(200)
larger_better = ["dm"]
def count_successes(df, accs, metric, algo, target, no_trials, pairs):
    count = 0
    for (x,y) in pairs:
        x_met = (df[(df["no"] == x) & (df["algo"] == algo)][metric])
        y_met = (df[(df["no"] == y) & (df["algo"] == algo)][metric])
        x_acc = (accs[(accs["no"] == x) & (accs["algo"] == algo)][target])
        y_acc = (accs[(accs["no"] == y) & (accs["algo"] == algo)][target])
        
        x_met = float(x_met)
        y_met = float(y_met)
        x_acc = float(x_acc)
        y_acc = float(y_acc)
    
        if ((x_met >= y_met and x_acc >= y_acc) | (x_met <= y_met and x_acc <= y_acc)):
            count += 1 if metric in larger_better else 0
        else:
            count += 1 if metric not in larger_better else 0
    return max(count, no_trials-count)
    
```

```python
no_trials = 1000
pairs = gencoordinates(200, no_trials)
results = {}
for (df,accs, name) in [(births, births_accs, "birth"), (rings, rings_accs, "ring"), 
                  (adults, adults_accs, "adult"), (hearts, hearts_accs, "heart")]:
    print(name)
    res = {}
    for metric in births.columns[2:]:
        print(metric)
        count_met = 0
        for algo in ["mondrian", "datafly", "datafly_shuffled"]:
            
            count_algo = 0
            
            for target in ["lr_acc", "lr_auroc", "rf_pca_acc", "rf_pca_auroc", "knn_pca_acc", "knn_pca_auroc"]:
                count_target = count_successes(df, accs, metric, algo, target, no_trials, pairs)
                #print(f"\t\t{target}: {count_target}")
                count_algo += count_target
                  
            print(f"\t{algo}: {count_algo}")
            count_met += count_algo
              
        perc = count_met / (3*6*no_trials)
        print(f"{metric}: {perc}")
        res[metric] = perc
    results[name] = res

results = pd.DataFrame.from_dict(results, orient="index")
results
```
