```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("ring_randoms/metrics.csv")
X = df.drop("lr_acc", axis=1)
X
```

```python
y = df["lr_acc"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train = X_train
train["lr_acc"] = y_train

train.to_csv("ring_randoms/trainset.csv", index=None)

test = X_test
test["lr_acc"] = y_test

test.to_csv("ring_randoms/testset.csv", index=None)
```
